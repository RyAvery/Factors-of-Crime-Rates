```python
# Add Matplotlib inline magic command
%matplotlib inline
%matplotlib notebook
# Dependencies and Setup
import os
import pandas as pd
from datetime import datetime as dt
import matplotlib.pyplot as plt
# Import the style from Matplotlib.
from matplotlib import style
# File to Load
san_francisco_pd_arrests_data_to_load = os.path.join('..', 'resources', 'San_Francisco_PD_Arrests_Data_2021.csv')
```

```python
# Reading San_Francisco_PD_Arrests_Data_2021.csv into a dataframe
sfpd_arrests_data_df = pd.read_csv(san_francisco_pd_arrests_data_to_load, encoding='cp1252')
pd.set_option('display.max_columns', None)
sfpd_arrests_data_df.head(10)
```

```python
sfpd_arrests_data_df.info()
```

```python
sfpd_arrests_data_df.count()
```

```python
# Formated object ['Report Datetime'] into datetime64 
sfpd_arrests_data_df['Report Datetime'] = pd.to_datetime(sfpd_arrests_data_df['Report Datetime'],
                                                         errors='coerce')
sfpd_arrests_data_df['Report Datetime']
```

```python
# Added new series to DataFrame by using list comprehension and zipping ['Latitude'] and ['Longitude'] into standard coordinate order ['Coordinates']
#x = [f'({a},{b})' for a, b in zip(df["a"], df["b"])]
sfpd_arrests_data_df['Coordinates'] = [f'({a},{b})' for a, b in zip(sfpd_arrests_data_df['Latitude'], sfpd_arrests_data_df['Longitude'])]
sfpd_arrests_data_df['Coordinates']
```

```python
# Created sorted DataFrame that contains all reports from 2021, also only picked the columns needed for analysis
sfpd_arrests_data_cleaned_df = sfpd_arrests_data_df.loc[(sfpd_arrests_data_df['Report Datetime']>='2021/01/01') & 
                                                        (sfpd_arrests_data_df['Report Datetime']<='2021/12/31'), 
                                                        ['Row ID', 'Report Datetime', 'Incident Subcategory',
                                                         'Incident Description', 'Analysis Neighborhood', 'Intersection', 
                                                         'Latitude', 'Longitude', 'Coordinates']].sort_values(by='Report Datetime').copy()
sfpd_arrests_data_cleaned_df.info()
```

```python
# Renaming columns to make more uniform
sfpd_arrests_data_cleaned_df = sfpd_arrests_data_cleaned_df.rename(columns={'Row ID':'Report ID',
                                                                      'Report Datetime':'Report Datetime',
                                                                      'Incident Subcategory':'Charge',
                                                                      'Incident Description':'Charge Description',
                                                                      'Analysis Neighborhood':'Neighborhood',
                                                                      'Intersection':'Address',
                                                                      'Latitude':'Lat',
                                                                      'Longitude':'Lon'})
```

```python
sfpd_arrests_data_cleaned_df.dropna(subset=['Charge','Charge Description'], thresh=2, inplace=True)
```

```python
# Reset index to 0 and drop old index column
sfpd_arrests_data_cleaned_df.reset_index(drop=True, inplace=True)
sfpd_arrests_data_cleaned_df.index.values
```

```python
# Double checking NaNs and column Dtype
sfpd_arrests_data_cleaned_df.info()
```

```python
# Print final DataFrame
pd.set_option('display.max_rows', None)
sfpd_arrests_data_cleaned_df.head(500)
```

```python
San_Francisco_PD_Arrests_Data_2021_clean_csv = os.path.join('..', 'resources', 'Cleaned', 'San_Francisco_PD_Arrest_Data_2021_clean.csv')
sfpd_arrests_data_cleaned_df.to_csv(San_Francisco_PD_Arrests_Data_2021_clean_csv, index=False, encoding='ISO=8859-1')
```
