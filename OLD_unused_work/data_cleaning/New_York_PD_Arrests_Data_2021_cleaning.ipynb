```python
# Add Matplotlib inline magic command
%matplotlib inline
%matplotlib notebook
# Dependencies and Setup
import os
import pandas as pd
from datetime import datetime as dt
import matplotlib.pyplot as plt
# Import the style from Matplotlib.
from matplotlib import style
# File to Load
new_york_pd_arrests_data_to_load = os.path.join('..', 'resources', 'New_York_PD_Arrests_Data_2021.csv')
```

```python
# New_York_PD_Arrests_Data_2021.csv into a dataframe
nypd_arrests_data_df = pd.read_csv(new_york_pd_arrests_data_to_load, encoding='cp1252')
pd.set_option('display.max_columns', None)
nypd_arrests_data_df.head(10)
```

```python
nypd_arrests_data_df.info()
```

```python
nypd_arrests_data_df.count()
```

```python
# Formated object ['ARREST_DATE'] into datetime64 
nypd_arrests_data_df['ARREST_DATE'] = pd.to_datetime(nypd_arrests_data_df['ARREST_DATE'],
                                                         errors='coerce')
nypd_arrests_data_df['ARREST_DATE']
```

```python
# Added new series to DataFrame by using list comprehension and zipping ['Latitude'] and ['Longitude'] into standard coordinate order ['Coordinates']
#x = [f'({a},{b})' for a, b in zip(df["a"], df["b"])]
nypd_arrests_data_df['Coordinates'] = [f'({a},{b})' for a, b in zip(nypd_arrests_data_df['Latitude'], nypd_arrests_data_df['Longitude'])]
nypd_arrests_data_df['Coordinates']
```

```python
# Created sorted DataFrame that contains all reports from 2021, also only picked the columns needed for analysis
nypd_arrests_data_cleaned_df = nypd_arrests_data_df.loc[(nypd_arrests_data_df['ARREST_DATE']>='2021/01/01') & 
                                                        (nypd_arrests_data_df['ARREST_DATE']<='2021/12/31'), 
                                                        ['ARREST_KEY', 'ARREST_DATE', 'AGE_GROUP', 'PERP_SEX', 
                                                         'PD_DESC', 'OFNS_DESC', 'Latitude', 'Longitude', 
                                                         'Coordinates']].sort_values(by='ARREST_DATE').copy()
nypd_arrests_data_cleaned_df.info()
```

```python
nypd_arrests_data_cleaned_df = nypd_arrests_data_cleaned_df.rename(columns={'Row ID':'Report ID',
                                                                      'Report Datetime':'Report Datetime',
                                                                      'Age':'Age'
                                                                      'Sex Code':'Sex',
                                                                      'Charge Group Description':'Charge',
                                                                      'Charge Description':'Charge Description',
                                                                      'Analysis Neighborhood':'Neighborhood',
                                                                      'Intersection':'Address',
                                                                      'Latitude':'Lat',
                                                                      'Longitude':'Lon'})
```
