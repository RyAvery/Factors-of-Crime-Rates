```python
# Add Matplotlib inline magic command
%matplotlib inline
%matplotlib notebook
# Dependencies and Setup
import os
import pandas as pd
from datetime import datetime as dt
import matplotlib.pyplot as plt
# Import the style from Matplotlib.
from matplotlib import style
# File to Load
seattle_pd_arrests_data_to_load = os.path.join('..', 'resources', 'Seattle_PD_Arrests_Data_2021.csv')
```

```python
# Reading Seattle_PD_Arrests_Data_2021.csv into a dataframe
spd_arrests_data_df = pd.read_csv(seattle_pd_arrests_data_to_load, encoding='cp1252')
pd.set_option('display.max_columns', None)
spd_arrests_data_df.head(10)
```

```python
spd_arrests_data_df.info()
```

```python
spd_arrests_data_df.count()
```

```python
# Formated object ['Report Datetime'] into datetime64 
spd_arrests_data_df['Report DateTime'] = pd.to_datetime(spd_arrests_data_df['Report DateTime'],
                                                         errors='coerce')
spd_arrests_data_df['Report DateTime']
```

```python
# Added new series to DataFrame by using list comprehension and zipping ['Latitude'] and ['Longitude'] into standard coordinate order ['Coordinates']
#x = [f'({a},{b})' for a, b in zip(df["a"], df["b"])]
spd_arrests_data_df['Coordinates'] = [f'({a},{b})' for a, b in zip(spd_arrests_data_df['Latitude'], spd_arrests_data_df['Longitude'])]
spd_arrests_data_df['Coordinates']
```

```python
# Created sorted DataFrame that contains all reports from 2021, also only picked the columns needed for analysis
spd_arrests_data_cleaned_df = spd_arrests_data_df.loc[(spd_arrests_data_df['Report DateTime']>='2021/01/01') & 
                                                      (spd_arrests_data_df['Report DateTime']<='2021/12/31'), 
                                                      ['Report Number', 'Report DateTime', 'Offense Parent Group', 
                                                       'Offense', 'MCPP', '100 Block Address', 'Latitude', 'Longitude', 
                                                       'Coordinates']].sort_values(by='Report DateTime').copy()
spd_arrests_data_cleaned_df.info()
```

```python
spd_arrests_data_cleaned_df = spd_arrests_data_cleaned_df.rename(columns={'Report Number':'Report ID',
                                                                      'Report Datetime':'Report Datetime',
                                                                      'Offense Parent Group':'Charge',
                                                                      'Offense':'Charge Description',
                                                                      'MCPP':'Neighborhood',
                                                                      '100 Block Address':'Address',
                                                                      'Latitude':'Lat',
                                                                      'Longitude':'Lon'})
```

```python
# Reset index to 0 and drop old index column
spd_arrests_data_cleaned_df.reset_index(drop=True, inplace=True)
spd_arrests_data_cleaned_df.index.values
```
